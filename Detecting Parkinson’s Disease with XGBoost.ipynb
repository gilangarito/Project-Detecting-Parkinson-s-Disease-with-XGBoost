<h2> Import Dataset </h2>

```python
# Import packages
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, plot_confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('max_columns', 100) # Show up to 100 columns
pd.set_option('max_rows', 200) # Show up to 100 columns
```

```python
# Import dataset
df = pd.read_csv('parkinsons.data')
df.head()
```

```python
# Data information
df.info()
```

```python
# Data describe
df.describe()
```

<h2> Checking missing values </h2>

```python
# Checking missing values
miss = df.isnull().sum()
miss_percnt = (miss/len(df))*100
pd.concat([miss,miss_percnt], axis=1, keys=['Missing Values','Missing Percentage'])
```

```python
# Get the count of each label (0 and 1) in labels
df['status'].value_counts()
```

Health status of the subject:
1 - Parkinson's;
0 - healthy

```python
# Separating the features and labels
features = df.loc[:,df.columns!='status'].values[:,1:]
labels = df['status'].values
```

<h2> Data Preprocessing </h2>

```python
# Scale the features to between -1 and 1
scaler = MinMaxScaler(feature_range=(-1, 1))
X = scaler.fit_transform(features)
y = labels
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

<h2> XGBoost </h2>

```python
# Train the model using XGBoost
model = XGBClassifier().fit(X_train, y_train)

# Predict the model
pred = model.predict(X_test)
```

```python
# Summary of the predictions made by the classifier
print(classification_report(y_test, pred))

# Visualize the Confusion Matrix
class_names = ['Healthy', 'Parkinson']
plot_confusion_matrix(model, X_test, y_test,  values_format = '.5g', display_labels=class_names)
plt.title("XGBoost Confusion Matrix")

# Accuracy score
print('Accuracy is',accuracy_score(pred, y_test).round(3))
```

XGClassifier model accuracy to detect the presence of parkinson disease is 94.9%.
